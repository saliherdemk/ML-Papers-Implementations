```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 500)

def energy(x):
    return x**2

temperatures = [0.1, 0.5, 1.0, 2.0, 5.0]

plt.figure(figsize=(10, 6))

for T in temperatures:
    unnormalized = np.exp(-energy(x) / T)
    Z = np.trapezoid(unnormalized, x)
    probs = unnormalized / Z

    plt.plot(x, probs, label=f"T = {T}")

plt.title("Boltzmann Distribution")
plt.xlabel("State (x)")
plt.ylabel("Probability Density P(x)")
plt.grid(True)
plt.legend()
plt.show()
```
